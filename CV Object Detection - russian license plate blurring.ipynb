# Computer Vision Object Detection

## russian license plate blurring

```python
#imports
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#data
img = cv2.imread('../Computer-Vision-with-Python/DATA/car_plate.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
def display(img):
    fig=plt.figure(figsize=(10,8))
    ax=fig.add_subplot(111)
    ax.imshow(img)
display(img)
plt.title('original image')

#bluring model
plate_cascade = cv2.CascadeClassifier('../Computer-Vision-with-Python/DATA/haarcascades/haarcascade_russian_plate_number.xml')
def detect_and_blur_plate(img):
    img_copy = img.copy()
    plate_rect = plate_cascade.detectMultiScale(img_copy,scaleFactor=1.2,minNeighbors=5)
    for (x,y,w,h) in plate_rect:
        roi = img_copy[y:y+h,x:x+w]
        roi_blur = cv2.medianBlur(roi,7)
        img_copy[y:y+h,x:x+w]=roi_blur
    return img_copy

#results
result = detect_and_blur_plate(img)
display(result)
plt.title('blurred plate')
```
